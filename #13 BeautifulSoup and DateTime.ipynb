```python
from bs4 import BeautifulSoup
```

```python
sample_data = """<html><head><title> this is my sample html page</title></head>
<body><h1>this is my heading</h1>
<h1>this is my heading</h1>
<h1>this is my heading</h1>
<h1>this is my heading</h1>
<h2>this is my heading</h2>
<h3>this is my heading</h3>
<h4>this is my heading</h4>
<h5>this is my heading</h5>
<p>my name is Vivek Kumar Verma</p>
</body>
</html>"""
```

```python
data_parse = BeautifulSoup(sample_data, 'html.parser')
```

```python
data_parse
```

```python
data_parse.title.string
```

```python
data_parse.h1.string
```

```python
data_parse.p.string
```

```python
for i in data_parse.find_all('h1'):
    print(i.string)
```
